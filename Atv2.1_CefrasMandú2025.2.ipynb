```python
def escalar_vetor(vetor, nova_magnitude):
    norma_original = np.linalg.norm(vetor)
    if norma_original == 0:
        print("Aviso: Não é possível escalar o vetor nulo.")
        return vetor
        
    vetor_unitario = vetor / norma_original
    
    vetor_escalado = vetor_unitario * nova_magnitude
    return vetor_escalado

v = np.array([3, 1, 5])
mag_desejada = 10

print(f"Vetor Original: {v}")
print(f"Norma Original: {np.linalg.norm(v):.4f}\n")

v_novo = escalar_vetor(v, mag_desejada)
print(f"Novo Vetor (magnitude {mag_desejada}): {v_novo}")
print(f"Norma do Novo Vetor: {np.linalg.norm(v_novo):.4f}")
```

```python
t = np.random.randn(2)
r = np.random.randn(2)

t_paralelo = r * (np.dot(t, r) / np.dot(r, r))

t_perp = t - t_paralelo

print(f"Vetor 't' original: {t}")
print(f"Soma dos componentes: {t_paralelo + t_perp}")
print(f"A soma é igual a 't'? {np.allclose(t, t_paralelo + t_perp)}\n")

dot_product = np.dot(t_perp, r)
print(f"Produto escalar entre t_perp e r: {dot_product:.15f}")
print(f"É ortogonal (produto escalar ~ 0)? {np.isclose(dot_product, 0)}\n")

plt.figure(figsize=(8, 8))

plt.arrow(0, 0, t[0], t[1], head_width=.08, width=.02, color='k', length_includes_head=True, label='t (alvo)')
plt.arrow(0, 0, r[0], r[1], head_width=.08, width=.02, color='gray', length_includes_head=True, label='r (referência)')
plt.arrow(0, 0, t_paralelo[0], t_paralelo[1], head_width=.08, width=.02, color='k', linestyle='--', length_includes_head=True, label='t_paralelo')

plt.arrow(t_paralelo[0], t_paralelo[1], t_perp[0], t_perp[1], head_width=.08, width=.02, color='k', linestyle=':', length_includes_head=True, label='t_perp')

plt.grid(linestyle='--')
plt.axis('equal')
plt.legend()
plt.title('Decomposição Ortogonal de t em relação a r')
plt.show()
```
